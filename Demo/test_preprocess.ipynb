```python
import numpy as np
import pandas as pd
import pickle

## note reflist.pkl is in the pwd in this example
## get reflist for AP dicrimination
with open('M2_reflist.pkl', 'rb') as f:
    reflist = pickle.load(f)

## base creates the base for the vector dataframe with correct APs
base = pd.read_pickle("M2_refbase.pkl")
```

```python
## process a vector for testing (at loc 1)
numdb = 20
locPath = "raw_data/loc_1_test/"
for k in range(1,numdb+1):
    dbPath = "db%s.csv" % k
    path = locPath + dbPath
    ## "file_{}.dat".format(i)
    data = pd.read_csv(path)
        
    new = data[['MAC Address','RSSI']]
    new = new.set_index('MAC Address')
    vecdf = base
    new = new['RSSI'].to_dict()
    applist = []
    for h in reflist:
        toapp = 0
        for x,y in new.items():
            if x == h:
                toapp = y
        applist.append(toapp)
    applist.append(3)
    add = np.asarray(applist)
    vecdf[('RSSI%d' % k)] = add
## vecdf now refers to all rssi values sorted by scan and AP(MAC Address) for 1 vector
## We must now transpose each df, find the average RSSI and append the resulting vector
## to our final data df, then export as a csv
vecdf = vecdf.T
## replace all zero values with nan
vecdf.replace(0, np.nan, inplace= True)
print(vecdf)

vecdf.to_csv('test1.csv', index=False)

## process a vector for testing (at loc 2)
numdb = 20
locPath = "raw_data/loc_2_test/"
for k in range(1,numdb+1):
    dbPath = "db%s.csv" % k
    path = locPath + dbPath
    ## "file_{}.dat".format(i)
    data = pd.read_csv(path)
        
    new = data[['MAC Address','RSSI']]
    new = new.set_index('MAC Address')
    vecdf = base
    new = new['RSSI'].to_dict()
    applist = []
    for h in reflist:
        toapp = 0
        for x,y in new.items():
            if x == h:
                toapp = y
        applist.append(toapp)
    applist.append(3)
    add = np.asarray(applist)
    vecdf[('RSSI%d' % k)] = add
## vecdf now refers to all rssi values sorted by scan and AP(MAC Address) for 1 vector
## We must now transpose each df, find the average RSSI and append the resulting vector
## to our final data df, then export as a csv
vecdf = vecdf.T
## replace all zero values with nan
vecdf.replace(0, np.nan, inplace= True)
print(vecdf)

vecdf.to_csv('test2.csv', index=False)

## process a vector for testing (at loc 3)
numdb = 20
locPath = "raw_data/loc_3_test/"
for k in range(1,numdb+1):
    dbPath = "db%s.csv" % k
    path = locPath + dbPath
    ## "file_{}.dat".format(i)
    data = pd.read_csv(path)
        
    new = data[['MAC Address','RSSI']]
    new = new.set_index('MAC Address')
    vecdf = base
    new = new['RSSI'].to_dict()
    applist = []
    for h in reflist:
        toapp = 0
        for x,y in new.items():
            if x == h:
                toapp = y
        applist.append(toapp)
    applist.append(3)
    add = np.asarray(applist)
    vecdf[('RSSI%d' % k)] = add
## vecdf now refers to all rssi values sorted by scan and AP(MAC Address) for 1 vector
## We must now transpose each df, find the average RSSI and append the resulting vector
## to our final data df, then export as a csv
vecdf = vecdf.T
## replace all zero values with nan
vecdf.replace(0, np.nan, inplace= True)
print(vecdf)

vecdf.to_csv('test3.csv', index=False)
```
