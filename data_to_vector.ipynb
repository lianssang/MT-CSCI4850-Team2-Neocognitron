```bash
## our data was taken continuosly from db1 - 450.csv for each loc
## so this bash file will sort it into vector sub folders
## additional cells below show other use cases

#!/bin/bash
for i in {1..3} ## number of locations
do
   let "n = 1" ## n will track the current dbXX.csv
   for j in {1..50} ## number of vectors desired
   do
       mkdir new_loc_$i/vec_$j
       for x in {1..9} ## number of scans in loc
       do
           mv new_loc_$i/db$n.csv new_loc_$i/vec_$j/db$n.csv
           let "n = n + 1"
       done
   done
   echo $i
   
done
```
